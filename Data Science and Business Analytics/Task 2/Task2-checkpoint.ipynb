### Unsupervised Clustering - Kmeans Clustering
### Divides data points into k distinct groups.

```python
x=data.iloc[:,[0,1,2,3]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
```

```python
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The elbow method')
plt.show()
```

### We choose our cluster value as the value where the slope is less steep. Here in this case, i'll be choosing value=3 since after 3, the drop is also minimal.

```python
kmeans=KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

### Visualizing the clusters once between sepallength and petal length 

```python
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s=50, c='red', label='Iris-setosa')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s=50, c='blue', label='Iris-versicolor')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1], s=50, c='green', label='Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', label='Centroids')
plt.legend()
```
